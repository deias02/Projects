#### Etapa 1: Identificar Valores Discrepantes

Um método simples para identificar outliers é usar o **IQR (Intervalo Interquartil)**, que se baseia nos quartis para detectar pontos fora do intervalo típico.

```python
# Função para identificar outliers com base no IQR (Quartis)
def detectar_outliers(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    print("limite inferior:", limite_inferior)
    print("limite superior:", limite_superior)
    
    # Filtrando os outliers
    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
    return outliers

# Detectando outliers na coluna 'lead_time' (exemplo)
outliers_lead_time = detectar_outliers(hotel_data1, 'lead_time')

# Exibir outliers detectados na coluna 'lead_time'
print("Número de linhas nos dados que contêm valores discrepantes em `lead_time`:", len(outliers_lead_time))
```

#### Etapa 4: Tratamento de Outliers

Após identificar os outliers, você pode optar por removê-los ou tratá-los com técnicas de imputação. Aqui estão duas abordagens:

```python
# Função para remover outliers de uma coluna
def remover_outliers(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # Remover outliers
    df_sem_outliers = df[(df[coluna] >= limite_inferior) & (df[coluna] <= limite_superior)]
    return df_sem_outliers

# Remover outliers da coluna 'lead_time'
hotel_data_sem_outliers = remover_outliers(hotel_data1, 'lead_time')
print(hotel_data_sem_outliers.describe())
```

O objetivo é comparar a performance dos diferentes modelos e escolher o melhor para prever se uma reserva será cancelada.

**1. Aprimorar o Modelo de Árvore de Decisão para Cancelamento**

Assim como fizemos para os pedidos especiais, ajustaremos os hiperparâmetros do modelo de árvore de decisão para prever cancelamentos, ajustando **max_depth** e **min_samples_leaf**.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Selecionando variáveis para o modelo preditivo de cancelamento
X = hotel_data[['lead_time', 'previous_cancellations', 'total_of_special_requests', 'required_car_parking_spaces']]
y = hotel_data['is_canceled']

# Dividindo o conjunto de dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Ajuste de hiperparâmetros para a árvore de decisão
modelo_cancelamento_aprimorado = DecisionTreeClassifier(random_state=42, max_depth=5, min_samples_leaf=10)

# Treinando o modelo
modelo_cancelamento_aprimorado.fit(X_train, y_train)

# Fazendo previsões com o conjunto de teste
y_pred_cancelamento_aprimorado = modelo_cancelamento_aprimorado.predict(X_test)

# Avaliando o novo modelo
acuracia_cancelamento_aprimorado = accuracy_score(y_test, y_pred_cancelamento_aprimorado)
print(f'Acurácia do Modelo Aprimorado (Cancelamento): {acuracia_cancelamento_aprimorado * 100:.2f}%')

# Matriz de confusão e relatório de classificação
matriz_confusao_cancelamento_aprimorado = confusion_matrix(y_test, y_pred_cancelamento_aprimorado)
print('Matriz de Confusão (Aprimorado - Cancelamento):')
print(matriz_confusao_cancelamento_aprimorado)

relatorio_cancelamento_aprimorado = classification_report(y_test, y_pred_cancelamento_aprimorado)
print('Relatório de Classificação (Aprimorado - Cancelamento):')
print(relatorio_cancelamento_aprimorado)
```

**2. Testar Algoritmo Random Forest para Cancelamento**

O Random Forest combina várias árvores de decisão e pode ser uma opção mais robusta. Vamos testar esse modelo para prever cancelamentos.

```python
from sklearn.ensemble import RandomForestClassifier

# Criando o modelo Random Forest
modelo_random_forest_cancelamento = RandomForestClassifier(n_estimators=100, random_state=42)

# Treinando o modelo Random Forest
modelo_random_forest_cancelamento.fit(X_train, y_train)

# Fazendo previsões com o conjunto de teste
y_pred_rf_cancelamento = modelo_random_forest_cancelamento.predict(X_test)

# Avaliando o modelo Random Forest
acuracia_rf_cancelamento = accuracy_score(y_test, y_pred_rf_cancelamento)
print(f'Acurácia do Modelo Random Forest (Cancelamento): {acuracia_rf_cancelamento * 100:.2f}%')

# Matriz de confusão e relatório de classificação
matriz_confusao_rf_cancelamento = confusion_matrix(y_test, y_pred_rf_cancelamento)
print('Matriz de Confusão (Random Forest - Cancelamento):')
print(matriz_confusao_rf_cancelamento)

relatorio_rf_cancelamento = classification_report(y_test, y_pred_rf_cancelamento)
print('Relatório de Classificação (Random Forest - Cancelamento):')
print(relatorio_rf_cancelamento)
```

**3. Testar Algoritmo K-Nearest Neighbors (KNN) para Cancelamento**

O K-Nearest Neighbors (KNN) também pode ser uma opção para prever cancelamentos, pois ele classifica as reservas com base nas reservas mais próximas em termos de características.


```python
from sklearn.neighbors import KNeighborsClassifier

# Criando o modelo K-Nearest Neighbors (KNN)
modelo_knn_cancelamento = KNeighborsClassifier(n_neighbors=5)

# Treinando o modelo KNN
modelo_knn_cancelamento.fit(X_train, y_train)

# Fazendo previsões com o conjunto de teste
y_pred_knn_cancelamento = modelo_knn_cancelamento.predict(X_test)

# Avaliando o modelo KNN
acuracia_knn_cancelamento = accuracy_score(y_test, y_pred_knn_cancelamento)
print(f'Acurácia do Modelo KNN (Cancelamento): {acuracia_knn_cancelamento * 100:.2f}%')

# Matriz de confusão e relatório de classificação
matriz_confusao_knn_cancelamento = confusion_matrix(y_test, y_pred_knn_cancelamento)
print('Matriz de Confusão (KNN - Cancelamento):')
print(matriz_confusao_knn_cancelamento)

relatorio_knn_cancelamento = classification_report(y_test, y_pred_knn_cancelamento)
print('Relatório de Classificação (KNN - Cancelamento):')
print(relatorio_knn_cancelamento)
```

**4. Comparação dos Modelos de Cancelamento**

Finalmente, podemos comparar os três modelos: **Árvore de Decisão Aprimorada, Random Forest e K-Nearest Neighbors (KNN)**.


```python
# Comparando as acurácias dos modelos para cancelamento
print(f'Acurácia Árvore de Decisão Aprimorada (Cancelamento): {acuracia_cancelamento_aprimorado * 100:.2f}%')
print(f'Acurácia Random Forest (Cancelamento): {acuracia_rf_cancelamento * 100:.2f}%')
print(f'Acurácia KNN (Cancelamento): {acuracia_knn_cancelamento * 100:.2f}%')
```
